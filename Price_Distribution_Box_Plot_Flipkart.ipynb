```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
data = pd.read_csv('/content/drive/MyDrive/flipkart_processed_output.csv')

# Convert "Price" column to numeric (if needed)
data["Price"] = pd.to_numeric(data["Price"], errors="coerce")

# Remove rows with NaN values in "Price" column
data = data.dropna(subset=["Price"])

# Create a histogram for the distribution of prices
plt.figure(figsize=(10, 6))
plt.hist(data["Price"], bins=20, color='blue', alpha=0.7)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Prices")
plt.tight_layout()

# Display the histogram
plt.show()

# Create box plots to observe price distributions by brand
plt.figure(figsize=(10, 6))
data.boxplot(column="Price", by="ProductName", vert=False)
plt.xlabel("Price")
plt.ylabel("Brand")
plt.title("Price Distribution by Brand")
plt.tight_layout()

# Display the box plots
plt.show()
```
