```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/amazon_processed_output.csv')

# Filter out non-numeric values from the 'Rating' column
numeric_ratings = pd.to_numeric(df['Rating'], errors='coerce')

# Remove rows with NaN values (caused by non-numeric values)
numeric_ratings = numeric_ratings.dropna()

# Convert 'Rating' column to integer
numeric_ratings = numeric_ratings.astype(int)

# Count the occurrences of each rating
rating_counts = numeric_ratings.value_counts()

# Remove '0' rating and ratings outside the range [1, 5]
rating_counts = rating_counts.loc[range(1, 6)]

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```
