```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
data = pd.read_csv('/content/drive/MyDrive/flipkart_processed_output.csv')

# Convert "Price" and "Rate" columns to numeric (if needed)
data["Price"] = pd.to_numeric(data["Price"], errors="coerce")
data["Rate"] = pd.to_numeric(data["Rate"], errors="coerce")

# Remove rows with NaN values in "Price" and "Rate" columns
data = data.dropna(subset=["Price", "Rate"])

# Create a scatter plot for Ratings vs. Price
plt.figure(figsize=(10, 6))
plt.scatter(data["Price"], data["Rate"], alpha=0.5)
plt.xlabel("Price")
plt.ylabel("Rating")
plt.title("Ratings vs. Price")
plt.tight_layout()

# Display the scatter plot
plt.show()
```
